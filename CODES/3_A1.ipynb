```python
#A1 Evaluate the intraclass spread and interclass distances between the classes in your dataset.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st

df = pd.read_excel('/content/TRAINING_DATA.xlsx')


class_labels = df.iloc[:, -1]

unique_labels = class_labels.unique()


class0_label = unique_labels[0]
class1_label = unique_labels[1]


class0_feats = df[class_labels == class0_label].iloc[:, :-1].values
class1_feats = df[class_labels == class1_label].iloc[:, :-1].values

# Calculate the mean vector (class centroid) for each class
class0_centroid = np.mean(class0_feats, axis=0)
class1_centroid = np.mean(class1_feats, axis=0)

# Calculate the intraclass spread (variance) for each class
class0_spread = np.var(class0_feats, axis=0)
class1_spread = np.var(class1_feats, axis=0)

# Calculate the interclass distance (Euclidean distance) between the two classes
interclass_distance = np.linalg.norm(class0_centroid - class1_centroid)

print("Class 0 Centroid\n", class0_centroid)
print("Class 1 Centroid\n", class1_centroid)
print("Class 0 Spread\n", class0_spread)
print("Class 1 Spread\n", class1_spread)
print("Interclass Distance\n", interclass_distance)
```

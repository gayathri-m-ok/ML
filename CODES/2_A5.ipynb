```python
""" A5.1 Study each attribute and associated values present. Identify the datatype (nominal etc.)
for the attribute """
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel("/content/Lab Session Data.xlsx", sheet_name ="thyroid0387_UCI")
print(df.head())

attribute_names = df.columns
print("Attribute Names:")
print(attribute_names)

data_types = df.dtypes
print("\nData Types:")
print(data_types)



numerical_attributes = df.select_dtypes(include=['int64', 'float64'])
summary_statistics = numerical_attributes.describe()
print("\nSummary Statistics for Numerical Attributes:")
print(summary_statistics)


categorical_attributes = df.select_dtypes(include=['object'])
for column in categorical_attributes.columns:
    unique_values = df[column].unique()
    value_counts = df[column].value_counts()
    print(f"\nUnique Values for '{column}':")
    print(unique_values)
    print(f"\nValue Counts for '{column}':")
    print(value_counts)
```

```python
# A2 Observe the density pattern for that feature by plotting the histogram.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
df = pd.read_excel("/content/TRAINING_DATA.xlsx")

feature = df.iloc[:, 0].values


hist, bins = np.histogram(feature, bins=10)

# Plot the histogram
plt.hist(bins[:-1], bins, weights=hist, edgecolor='black')
plt.xlabel('Feature Values')
plt.ylabel('Frequency')
plt.title('Histogram of Feature 1')
plt.show()

# Calculate the mean of the feature
feature_mean = np.mean(feature)
print("Mean of the Feature 1 \n", feature_mean)

# Calculate the variance of the feature
feature_variance = np.var(feature)
print("Variance of Feature 1 \n", feature_variance)
```

```python
# A1.4 Pseudo-Inverse find the cost of each product available for sale
import pandas as pd
import numpy as np
df = pd.read_excel("/content/purchase_data.xlsx")
pseudo_inverse = np.linalg.pinv(A)
print("Pseudo-Inverse of matrix A \n")
print(pseudo_inverse)
A_inv = np.linalg.pinv(A)
product_costs = np.matmul(A_inv, C)
print("Cost of each product available for sale \n")
print(product_costs)
```

```python
# Using elbow plot, determine the optimal k value for k-means clustering.
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import seaborn as sns


df = pd.read_excel('/content/TRAINING_DATA.xlsx')

# Converting the DataFrame to a NumPy array
X_train = df.values
distortions = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k).fit(X_train)
    distortions.append(kmeans.inertia_)

# Ploting the elbow plot
plt.plot(range(2, 20), distortions, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion (Sum of Squared Errors)')
plt.title('Elbow Plot for Optimal k Value')
plt.show()
```
